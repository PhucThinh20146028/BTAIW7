```python
import numpy as np
import pandas as pd
url = '/content/abalone.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)

sns.pairplot(data)


#am la ti le nghich
#duong la ti le thuan
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind = 'scatter', x = 'Length', y = 'Rings', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()

model.fit(X = pd.DataFrame(data['Length']), y = data['Rings'])
print('Gia tri Intercept la: ', model.intercept_)
print('Gia tri Coeffient la: ', model.coef_)
model.score(X = pd.DataFrame(data['Length']), y = data['Rings'])
```

```python
data.plot(kind = 'scatter', x = 'Length', y = 'Rings', figsize = (9,9), color = 'black')
y_pred = model.predict(X = pd.DataFrame(data['Length']))
plt.plot(data['Length'], y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Length'],data['Length']**2]).T

y = data['Rings']

model.fit(X,y = data['Rings'])
print('He so Intercept la: ', model.intercept_)

print('He so Coeffient la: ', model.coef_)
print('Do chinh xac cua mo hinh la: ', model.score(X,y))
```

```python
range = np.arange(0,1,0.1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'Length', y = 'Rings', figsize = (9,9), color = 'black')
plt.plot(range,y_pred,color='blue')
```
